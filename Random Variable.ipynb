## Probability Calculation for a Continous Random Variable

### The length of a human pregnancy is normally distributed with a mean of 272 days with a standard deviation of 9 days.

1. State the random variable.
2. Find the probability of a pregnancy lasting more than 280 days.
3. Find the probability of a pregnancy lasting less than 250 days.
4. Find the probability that a pregnancy lasts between 265 and 280 days.

### Solution

#### 1. RV - Length of Human Pregnancy

```python
from scipy import stats
```

#### 2.Find the probability of a pregnancy lasting more than 280 days.

```python
1 - stats.norm.cdf(x = 280, loc = 272, scale = 9)
```

#### 3.Find the probability of a pregnancy lasting less than 250 days.

```python
stats.norm.cdf(x = 250, loc = 272, scale = 9)
```

#### 4.Find the probability that a pregnancy lasts between 265 and 280 days.

```python
stats.norm.cdf(x = 280 , loc = 272, scale = 9) - stats .norm.cdf(x = 265, loc = 272, scale = 9)
```
