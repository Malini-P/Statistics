# Statistical Tests Tasks

## Question 1:

#### The historical average score in a certification exam is 65. A training institute introduces a new course. A random sample of 50 students trained under the new course has a mean score of 68 and a standard deviation of 10. At the 5% level of significance, test whether the new course improves performance.

```python
# H0: Current Process/Statement of No Change - There is no significant difference in the test scores.
# H1: Statement by change/New process - There is a significant difference in the average new course test scores.
```

```python
from scipy import stats
import numpy as np

observations_50_students = np.random.normal(loc = 68, scale=5.0, size = 50).astype('int')
t_stats,p_value = stats.ttest_1samp( a = observations_50_students , popmean = 65, alternative = 'greater')

p_value = p_value/2

print("t-statistic:", t_stats)
print("p-value:", p_value)

if p_value < 0.05:
    print("We Reject H0")
else:
    print("We do not Reject H0")
```

### At 5% level of significance, we can reject the Null Hypothesis and claim that the new course improves performance.

## Question 2:

#### A company claims that the average time taken to resolve customer tickets is 30 minutes. A random sample of 40 tickets shows a mean resolution time of 33 minutes. At the 5% level of significance, test whether the average resolution time has increases.

```python
# H0: Current process/Statement of No change - There is no significant difference in the average resolution time.
# H1: Statement by change/New process - There is a significant difference in the average resolution time.
```

```python
from scipy import stats
import numpy as np

observations_40_tickets = np.random.normal(loc = 33 , scale = 5.0, size = 40).astype("int")
t_stats,p_value = stats.ttest_1samp(a = observations_40_tickets, popmean = 30 , alternative = 'greater')

p_value = p_value/2

print("t-statistic:", t_stats)
print("p-value:", p_value)

if p_value < 0.05:
    print("We Reject H0")
else:
    print("We do not Reject H0")
```

### At 5% level of significance, we can reject the Null Hypothesis and claim that the average resolution time has increased.

## Question 3:

#### A company claims that 60% of its users are satisfied with its product. In a survey of 200 users, 134 users report satisfaction. At the 5% level of significance, test whether customer satisfaction is higher than claimed.

```python
# H0: Current process/Statement of No change - There is no significant difference in the survey report.
# H1: Statement by change/New process - There is a significant difference in the satisfaction survey report for the product. 
```

```python
pip install statsmodels
```

```python
from statsmodels.stats.proportion import proportions_ztest

# Given data
satisfied_users = 134
total_users = 200
p0 = 0.60   # claimed proportion

# One-sample proportion Z-test (right-tailed)
z_stat, p_value = proportions_ztest( count=satisfied_users, nobs=total_users, value=p0, alternative='larger')

# Decision
if p_value < 0.05:
    print("We Reject H0")
else:
    print("We do not Reject H0")

print("Z-statistic:", z_stat)
print("p-value:", p_value)
```

### At 5% level of significance, we can reject the Null Hypothesis and consider that the customer satisfaction is higher than claimed.

## Question 5:

#### Two teaching methods are used to train employees. A sample of 30 employees trained using Method A has a mean score of 75 with a standard deviation of 12. A sample of 28 employees trained using Method B has a mean score of 70 with a standard deviation of 10. At the 5% level of significance, test whether there is a difference in average scores between the two methods

```python
import numpy as np
from scipy import stats

# Simulated data based on given values
method_A = np.random.normal(loc=75, scale=12, size=30)
method_B = np.random.normal(loc=70, scale=10, size=28)

# Two-sample t-test (independent, two-tailed)
t_stat, p_value = stats.ttest_ind(a = method_A, b = method_B, equal_var=False)   # safer (Welch’s t-test)

# Decision
if p_value < 0.05:
    print("We Reject H0")
else:
    print("We do not Reject H0")

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

### At 5% level of significance, we can reject the Null Hypothesis and consider that there is a difference in average scores between the two methods

## Question 6:

#### The average delivery time from two warehouses is compared using random samples. Warehouse A has a sample of 35 deliveries with a mean time of 48 minutes and a standard deviation of 6 minutes. Warehouse B has a sample of 40 deliveries with a mean time of 52 minutes and a standard deviation of 7 minutes. At the 5% level of significance, test whether the mean delivery times differ.

```python
import numpy as np
from scipy import stats

# Set seed → ensures same output every time
#np.random.seed(0)

# Generate sample data based on given stats
method_A = np.random.normal(loc=48, scale=6, size=35)
method_B = np.random.normal(loc=52, scale=7, size=40)

# Two-sample t-test (Welch’s test)
t_stat, p_value = stats.ttest_ind( a = method_A, b = method_B, equal_var=False)

# Decision
if p_value < 0.05:
    print("We Reject H0")
else:
    print("We do not Reject H0")

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

### At the 5% level of significance, we reject the null hypothesis and conclude that there is a significant difference in delivery times between the two warehouses.

## Question 7:

#### Two versions of a website are tested. Out of 500 users for Version A, 320 users click the signup button. Out of 480 users for Version B, 290 users click the signup button. At the 5% level of significance, test whether the conversion rates differ.

```python
from statsmodels.stats.proportion import proportions_ztest

# Given data
clicks = [320, 290]      # successes
users = [500, 480]      # total users

# Two-sample proportion test (two-tailed)
z_stat, p_value = proportions_ztest( count=clicks, nobs=users, alternative='two-sided')

# Decision
if p_value < 0.05:
    print("We Reject H0")
else:
    print("We do not Reject H0")

print("Z-statistic:", z_stat)
print("p-value:", p_value)
```

#### At the 5% level of significance, we do not reject the null hypothesis and conclude that the conversion rates of the two website versions do not differ significantly.

## Question 8:

#### A vaccine is tested in two regions. In Region A, 180 out of 300 people recover. In Region B, 165 out of 310 people recover. At the 5% level of significance, test whether the recovery rates are significantly different.

```python
from statsmodels.stats.proportion import proportions_ztest

# Given data
recoveries = [180, 165]   # successes
total = [300, 310]        # total people

# Two-sample proportion test (two-tailed)
z_stat, p_value = proportions_ztest(count=recoveries, nobs=total, alternative='two-sided')

# Decision
if p_value < 0.05:
    print("We Reject H0")
else:
    print("We do not Reject H0")

print("Z-statistic:", z_stat)
print("p-value:", p_value)
```

## Question 9:

#### A company wants to compare the average productivity of employees trained using three different training programs. Program A has 25 employees with a mean productivity score of 80 and a standard deviation of 8. Program B has 30 employees with a mean of 75 and a standard deviation of 7. Program C has 28 employees with a mean of 82 and a standard deviation of 6. At the 5% level of significance, test whether there is a difference in mean productivity across the programs.

```python
#H₀: μ₁ = μ₂ = μ₃ (all means equal)
#H₁: At least one mean is different

#Interpretation:
#If F ≈ 1 → groups are similar →  don’t reject
#If F is large → groups are very different → reject 
```

```python
import numpy as np
from scipy import stats

# Fix randomness → same result every time
np.random.seed(0)

# Generate sample data
program_A = np.random.normal(loc=80, scale=8, size=25)
program_B = np.random.normal(loc=75, scale=7, size=30)
program_C = np.random.normal(loc=82, scale=6, size=28)

# One-way ANOVA
f_stat, p_value = stats.f_oneway(program_A, program_B, program_C)

# Decision
if p_value < 0.05:
    print("We Reject H0")
else:
    print("We do not Reject H0")

print("F-statistic:", f_stat)
print("p-value:", p_value)
```

### At the 5% level of significance, we reject the null hypothesis and conclude that at least one training program has a different mean productivity.

## Question 10:

#### A company wants to test whether gender and product preference are independent. The survey results are as follows:
1. Male: 60 prefer Product A, 40 prefer Product B
2. Female: 50 prefer Product A, 50 prefer Product B

At the 5% level of significance, test for independence.


```python
#H₀: Gender and product preference are independent
#H₁: They are not independent (there is an association)
```

```python
import numpy as np
from scipy.stats import chi2_contingency

# Contingency table
# Rows: Male, Female
# Columns: Product A, Product B
data = np.array([[60, 40],[50, 50]])

# Chi-square test
chi2, p_value, dof, expected = chi2_contingency(data)

# Decision
if p_value < 0.05:
    print("We Reject H0")
else:
    print("We do not Reject H0")

print("Chi-square statistic:", chi2)
print("p-value:", p_value)
print("Degrees of freedom:", dof)
```

### At the 5% level of significance, we fail to reject the null hypothesis and conclude that gender and product preference are independent.

## Question 11:

#### A  hospital studies whether smoking status is related to disease occurrence. The data collected is as follows:
1. Smokers: 70 have the disease, 30 do not
2. Non-smokers: 40 have the disease, 60 do not.

At the 5% level of significance, test whether smoking status and disease occurrence are independent. 


```python
import numpy as np
from scipy.stats import chi2_contingency

# Contingency table
# Rows: Smokers, Non-smokers
# Columns: Disease, No Disease
data = np.array([[70, 30],[40, 60]])

# Chi-square test
chi2, p_value, dof, expected = chi2_contingency(data)

# Decision
if p_value < 0.05:
    print("We Reject H0")
else:
    print("We do not Reject H0")

print("Chi-square statistic:", chi2)
print("p-value:", p_value)
print("Degrees of freedom:", dof)
```

## Question 12:

#### 12.	A dice manufacturer claims that a die is fair. The die is rolled 120 times, and the observed frequencies are:
1. Face 1: 18
2. Face 2: 22
3. Face 3: 20
4. Face 4: 25
5. Face 5: 17
6. Face 6: 18

At the 5% level of significance, test whether the die is fair.


```python
import numpy as np
from scipy.stats import chisquare

# Observed frequencies
observed = np.array([18, 22, 20, 25, 17, 18])

# Expected frequencies (fair die)
expected = np.array([20, 20, 20, 20, 20, 20])

# Chi-square goodness-of-fit test
chi2, p_value = chisquare(f_obs=observed, f_exp=expected)

# Decision
if p_value < 0.05:
    print("We Reject H0")
else:
    print("We do not Reject H0")

print("Chi-square statistic:", chi2)
print("p-value:", p_value)
```

### At the 5% level of significance, we fail to reject the null hypothesis and conclude that the die is fair.
